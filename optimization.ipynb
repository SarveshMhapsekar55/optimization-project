# Optimization Model - Profit Maximization

## Problem Statement

A company produces two products: Product A and Product B.

Each product requires limited resources such as labor and material.

The objective is to determine how many units of each product should be produced in order to maximize total profit while satisfying resource constraints.

## Objective Function

Maximize total profit:

Profit = 20x + 30y

Where:
- x = number of units of Product A
- y = number of units of Product B

## Constraints

The production is limited by the following constraints:

- Labor Constraint:  
  2x + y ≤ 40

- Material Constraint:  
  x + 2y ≤ 50

- Non-negativity Constraints:  
  x ≥ 0, y ≥ 0

```python
from pulp import *

# Create problem
model = LpProblem("Maximize_Profit", LpMaximize)

# Variables
x = LpVariable('Product_A', lowBound=0)
y = LpVariable('Product_B', lowBound=0)

# Objective function
model += 20*x + 30*y

# Constraints
model += 2*x + y <= 40
model += x + 2*y <= 50

# Solve
model.solve()

# Output
print("Status:", LpStatus[model.status])
print("Product A:", x.varValue)
print("Product B:", y.varValue)
print("Total Profit:", value(model.objective))
```

## Conclusion

The optimal production plan suggests producing:

- 10 units of Product A  
- 20 units of Product B  

This results in a maximum profit of 800.

This model demonstrates how optimization techniques can help businesses efficiently allocate resources and maximize profit.
